```python
def checklens(key_names,r):
    """
    Checks if keys in recipes are truly items and not a typo error
    """
    n=len(key_names)
    ok=True
    for i in range(len(r)):
        ok=(ok and len(r[i])==n)
    return ok
```

```python
##Pretty print functions
def statepathdict(x,r,stateini,key_names):
    statefi=stateini.copy()
    for i in range(len(x)):
        k=int(x[i])
        if k:
            for j in range(len(key_names)):
                statefi[key_names[j]]+=k*r[i][key_names[j]]
    return statefi
def ppr(r,key_names):
    st=""
    for i in range(len(key_names)):
        if r[key_names[i]]<0:
            st+=str(r[key_names[i]])+'*'+str(key_names[i])+' '
    st+='--> '
    for i in range(len(key_names)):
        if r[key_names[i]]>0:
            st+=str(r[key_names[i]])+'*'+str(key_names[i])+' '
    print(st)
def ppd(d):
    st="You will get : \n"
    newDict = dict(filter(lambda elem: elem[1] > 0, d.items()))
    for (key,val) in newDict.items():
        st+="  "+str(val)+" * "+str(key)+'\n'
    print(st)
def ppi(d):
    st="\n Starting with : \n"
    newDict = dict(filter(lambda elem: elem[1] > 0, d.items()))
    for (key,val) in newDict.items():
        st+="  "+str(val)+" * "+str(key)+'\n'
    print(st)
def pps(d):
    st="Scored by : \n"
    newDict = dict(filter(lambda elem: elem[1] > 0, d.items()))
    for (key,val) in newDict.items():
        st+="  "+str(key)+' = '+str(val)+'\n'
    print(st)
```

```python
def main3(r,key_names,resini,scores):
    # Create the mip solver with the CBC backend.
    solver = pywraplp.Solver('simple_mip_program',
                         pywraplp.Solver.CBC_MIXED_INTEGER_PROGRAMMING)
    
    infinity = solver.infinity()
    x = {}
    valmax=20000 #Arbitrary value, inifinty causes the problem to be infeasible, don't know why...
    
    #Non-zero initial ressources
    nonzerres=[]
    for i in range(len(key_names)):
        if resini[key_names[i]]!=0:
            nonzerres.append(key_names[i])
            
    #Non-zero score values
    nonzersc=[]
    for i in range(len(key_names)):
        if scores[key_names[i]]!=0:
            nonzersc.append(key_names[i])
    #Array of values to optimize (x[i] is the number of times recipe i is used
    
    for j in range(len(r)):
        x[j] = solver.IntVar(0, valmax, 'x[%i]' % j)

#   print('Number of variables =', solver.NumVariables())
    for i in range(len(key_names)):
        if key_names[i] in nonzerres:
            constraint = solver.RowConstraint(-resini[key_names[i]], 0, '')
        else:
            constraint = solver.RowConstraint(0, valmax, '')

        for j in range(len(r)):
            constraint.SetCoefficient(x[j], r[j][key_names[i]])
    print('Number of constraints =', solver.NumConstraints())
    
    sc=0
    for i in range(len(nonzersc)):
        for j in range(len(r)):
            if r[j][nonzersc[i]]!=0:
                sc+=scores[nonzersc[i]]*r[j][nonzersc[i]]*x[j]
    solver.Maximize(sc)

    status = solver.Solve()
    if status == pywraplp.Solver.OPTIMAL:
        res=[]
        print('Objective value =', solver.Objective().Value())
        for j in range(len(r)):
            ppr(r[j],key_names)
            print(x[j].name(), ' = ', x[j].solution_value())
            res.append(x[j].solution_value())
            print()
            
        print('Problem solved in %f milliseconds' % solver.wall_time())
        print('Problem solved in %d iterations' % solver.iterations())
        print('Problem solved in %d branch-and-bound nodes' % solver.nodes())
        #print(res)
        ppi(resini)
        pps(scores)
        ppd(statepathdict(res,r,resini,key_names))
        return(res)
        
    else:
        print('The problem does not have an optimal solution.')
```
